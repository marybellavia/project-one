```python
# importing dependencies
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# bringing in the datasets
food_data_path = "Resources/FoodDesert2015.csv"
stores_data_path = "Resources/StoresData.csv"
restaurant_data_path = "Resources/RestaurantData.csv"

food_df = pd.read_csv(food_data_path)
stores_df = pd.read_csv(stores_data_path)
rest_df = pd.read_csv(restaurant_data_path)
```

```python
# dropping rows with null values and confirming count
food_df.dropna()
food_df.count()
```

```python
#filter dataframe by Missouri
missouri_df = food_df[food_df["State"] == "Missouri"]
missouri_df
```

```python
# saving the df as a csv
missouri_df.to_csv("Resources/Missouri.csv")
```

```python
# looking at the dataframe
stores_df
```

```python
# doing a count
stores_df.count()
```

```python
# dropping rows with null values and confirming count
stores_df.dropna()
stores_df.count()
```

```python
# filtering down to Missouri
mo_stores_df = stores_df[stores_df["State"] == "MO"]
mo_stores_df
```

```python
# creating df with only relevant rows
reduced_stores_df = stores_df[['State',
                               'County',
                               'GROC09',
                               'SUPERC09',
                               'CONVS09',
                               'SPECS09'
                              ]]
# renaming columns
reduced_stores_df = reduced_stores_df.rename(columns={"GROC09": "GroceryStores",
                                                      "SUPERC09": "SupercentersClubStores",
                                                      "CONVS09": "ConvenienceStores",
                                                      "SPECS09" : "SpecializedFoodStores"
                                                     })
reduced_stores_df.head(20)
```

```python
rest_df
```

```python
# doing a count
rest_df.count()
```

```python
# dropping rows with null values and confirming count
rest_df.dropna()
rest_df.count()
```

```python
# filtering down to Missouri
mo_rest_df = rest_df[rest_df["State"] == "MO"]
mo_rest_df
```

```python
# creating df with only relevant rows
reduced_rest_df = rest_df[['State',
                           'County',
                           'FFR09',
                           'FSR09'
                          ]]
# renaming columns
reduced_rest_df = reduced_rest_df.rename(columns={"FFR09": "FastFoodRestaurants",
                                                  "FSR09": "FullServiceRestaurants"
                                                 })

reduced_rest_df.head(20)
```

```python
# creating a new df with all the info by county for grocery stores and restaurants in Missouri
all_food_df = pd.DataFrame({"State": reduced_rest_df['State'],
                            "County": reduced_rest_df['County'],
                            "GroceryStores": reduced_stores_df['GroceryStores'],
                            "SupercentersClubStores": reduced_stores_df['SupercentersClubStores'],
                            "ConvenienceStores": reduced_stores_df['ConvenienceStores'],
                            "SpecializedFoodStores": reduced_stores_df['SpecializedFoodStores'],
                            "FastFoodRestaurants": reduced_rest_df['FastFoodRestaurants'],
                            "FullServiceRestaurants": reduced_rest_df['FullServiceRestaurants']
                           })
all_food_df
```

```python
# saving the df as a csv
all_food_df.to_csv("Resources/StoresRestaurantData.csv")
```
